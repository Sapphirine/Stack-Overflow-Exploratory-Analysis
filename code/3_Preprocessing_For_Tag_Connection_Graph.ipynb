```python
import numpy as np
import pandas as pd
import collections
import matplotlib.pyplot as plt
```

```python
t=pd.read_csv('Tags.csv')
at256 = pd.read_csv('256tagappendforquestion.csv')
at256 = at256[at256.sumtag != 0]
```

```python
#Construct a co-occurence matrix for the tag graph
n=256
tagCount =  collections.Counter(list(t['Tag'])).most_common(n)
ocurrence = at256.iloc[:,2:n+2].copy()
co_ocurrence=np.dot(ocurrence.transpose(), ocurrence)
for i in range(n):
    co_ocurrence[i,i]=0
```

```python
#Construct the observed-to-expected table for the tag graph
co_matrix1 = np.empty(shape=((n*n-n)//2,2), dtype='object')
co_matrix2 = np.zeros(shape=((n*n-n)//2,5))
countTotal=sum(sum(co_ocurrence))/2
tagCount=np.array(tagCount)
iter=0
for i in range(n):
    for j in range(i+1,n):
        co_matrix1[iter, 0]=tagCount[i, 0]
        co_matrix1[iter, 1]=tagCount[j, 0]
        co_matrix2[iter, 1]=tagCount[i, 1]
        co_matrix2[iter, 2]=tagCount[j, 1]
        co_matrix2[iter, 3]=co_ocurrence[i, j]
        co_matrix2[iter, 4]=countTotal
        co_matrix2[iter, 0]=co_matrix2[iter, 3]*countTotal/co_matrix2[iter, 1]/co_matrix2[iter, 2]
        iter=iter+1
co_matrix=np.hstack((co_matrix1, co_matrix2))
co_ocurrencepd=pd.DataFrame(co_ocurrence)
co_ocurrencepd.head(3)
```

```python
co_matrix=pd.DataFrame(co_matrix)
co_matrix.head(3)
```

```python
cccc=co_matrix
#eliminating tag pairs that do not co-occur
cccc = cccc[~(cccc == 0).any(axis=1)]
sortedco_matrix=cccc.sort_values(by=2, ascending=1)
sortedco_matrix.columns = ['Tag1', 'Tag2', 'OERatio','Count1','Count2','Count12','CountAll']
```
